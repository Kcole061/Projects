```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read in Data
df = pd.read_csv(r'C:\Users\kahli\Desktop\Projects\movies.csv')
```

```python
df.head(10)
```

```python
# Checking the date

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.isnull().sum() 
```

```python
df.dtypes
```

```python
df = df.fillna(0)
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')


```

```python
df.head()
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].astype(str).str[-20:-16]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Scatter plot with budget vs gross
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.scatter(x=df['budget'], y = df['gross'])
plt.show()
```

```python
# Plot budget vs gross in seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Checking Correlation
df.corr()
# budget gross correlation is positive 0.75
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Corr Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

        
df_numerized.head()        
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Corr Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```
